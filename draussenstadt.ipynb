```python
title = []
date_time = []
location = []
link = []

#num_iter = len(soup.select('div > label'))

for i in range(50):
    title.append(soup.select('#e'+str(i)+'> div > div.col-sm-6.col-md-4 > strong > span')[0].get_text())
    location.append(soup.select('#e'+str(i)+'> div > div:nth-child(7)')[0].get_text())
    date_time.append(soup.select('#e'+str(i)+'> div > div:nth-child(4)')[0].get_text())
    link.append('https://www.smb.museum'+soup.select('#e'+str(i)+'> div > div.col-md-2.image > a')[0]['href'])
```

```python
table_info = pd.DataFrame({'Event Title':title, 'Location':location, 'Date and Time':date_time, 'Event Link':link})
```

```python
# Function to extract day, date, and hour from a single string
def extract_info(string):
    day = string[25:28]
    date = string[29:39]
    hour = string[39:45]
    return day, date, hour

# Apply the function to the 'date and time' column to create separate columns for day, date, and hour
table_info1[['Day', 'Date', 'Hour']] = table_info1['Date and Time'].apply(extract_info).apply(pd.Series)

# Drop the original 'date and time' column if needed
table_info1.drop(columns=['Date and Time'], inplace=True)

print(table_info1)





```
