```python

from bs4 import BeautifulSoup
import pandas as pd


# Find all elements with class "basic list-item today"
events = soup.find_all(class_='basic list-item today')

# Initialize lists to store the event details
titles = []
categories = []
organizers = []
venues = []
start_dates = []

# Extract event details and store in lists
for event in events:
    title = event.find('dsb-card-event')['text-title']
    category = event.find('dsb-card-event')['text-category']
    organizer = event.find('dsb-card-event')['text-organizer']
    venue = event.find('dsb-card-event')['text-venue']
    start_date = event.find('dsb-card-event')['start-date']
    titles.append(title)
    categories.append(category)
    organizers.append(organizer)
    venues.append(venue)
    start_dates.append(start_date)

# Create a DataFrame using Pandas
data = {
    'Title': titles,
    'Category': categories,
    'Organizer': organizers,
    'Venue': venues,
    'Start Date': start_dates
}
df = pd.DataFrame(data)

print(df)



```

```python

# Split the "Start Date" column into separate "Date" and "Hour" columns
df[['Date', 'Hour']] = df['Start Date'].str.split('T', expand=True)


print(df)
```
