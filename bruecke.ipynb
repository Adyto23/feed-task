```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Replace 'your_url_here' with the actual URL of the page containing events
url = "https://www.bruecke-museum.de/en/programm/veranstaltungen"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Fetch all event elements
events = soup.select('div.preview__info-column')

# Create a list to store dictionaries for each event
events_list = []

# Loop through each event and extract the information
for event in events:
    title = event.select_one('h1 > a').get_text()
    date_time = event.select_one('div:nth-child(3) > p').get_text()
    desc = event.select_one('div.preview__info-column > h2').get_text()
    
    try:
        Price = event.select_one('div:nth-child(5) > p').get_text()
    except AttributeError:
        price = "Category Not Available"
    
    link = f"https://www.bruecke-museum.de{event.select_one('h1 > a')['href']}"
    
    # Append event details to the list
    events_list.append({
        'Title': title,
        'Date_Time': date_time,
        'Description': desc,
        'Link': link,
    })

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(events_list)

# Print the DataFrame
print(df)
```

```python
import pandas as pd
import re

# Assuming you already have the DataFrame 'df' with the scraped data

# Clean the 'Date_Time' column
df['Date_Time'] = df['Date_Time'].apply(lambda date_time: re.sub(r'[^\d.\w\s,:]', '', date_time).strip())

# Extract 'Date' and 'Time' using regular expressions
date_time_pattern = r'(\d{1,2}\.\s\w+\s\d{4}),\s*(every\sSunday, noon|[\d.:apm]+\s*-\s*[\d.:apm]+|\w+\s*\w*)'
df[['Date', 'Time']] = df['Date_Time'].str.extract(date_time_pattern)

# Clean up the 'Date' and 'Time' columns
df['Date'] = df['Date'].str.strip()
df['Time'] = df['Time'].str.strip()


# Print the cleaned DataFrame
print(df)



```
