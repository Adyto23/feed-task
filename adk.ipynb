```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Replace 'url' with the actual URL you want to scrape
url = "https://www.adk.de/en/programme/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Fetch all elements for date, title, description, type, and link
dates = soup.select('div.event-info-date')
soup.select('div.event-text > h2 > a')
descriptions = soup.select('div.event-text > p:nth-child(2)')
event_types = soup.select('div.event-info > div.rubrik')
infos = soup.select('div.event-info')
languages = soup.select('p:nth-child(4)')
prices = soup.select('p:nth-child(5)')

# Ensure all lists have the same length
num_events = min(len(dates), len(titles), len(descriptions), len(event_types), len(infos), len(languages), len(prices))

# Create a list to store dictionaries for each event
events_list = []

# Loop through each event and extract the information
for i in range(num_events):
    # Extract the link    
    link = soup.select('div.event-info > div.ics-file > a')[i]['href'] if i < len(soup.select('div.event-info > div.ics-file > a')) else None

    # Clean up the text in all columns
    date = dates[i].get_text().strip() if i < len(dates) else None
    title = titles[i].get_text().strip() if i < len(titles) else None
    description = descriptions[i].get_text().strip() if i < len(descriptions) else None
    event_type = event_types[i].get_text().strip() if i < len(event_types) else None
    info = infos[i].get_text().strip() if i < len(infos) else None
    language = languages[i].get_text().strip() if i < len(languages) else None
    price = prices[i].get_text().strip() if i < len(prices) else None

    event = {
        'Date': date,
        'Title': title,
        'Description': description,
        'Event_Type': event_type,
        'Info': info,
        'Language': language,
        'Price': price,
        'Link': link
    }
    events_list.append(event)

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(events_list)

# Print the DataFrame
print(df)
```
